```python
import pandas as pd
```

```python
STATION_COLUMNS = ['Water Type', 'Elevation', 'Latitude', 'Longitude']
WATER_TYPES = ['Lake station']

df = pd.read_csv('data/GFQA_v2/GEMStat_station_metadata.csv',
                 encoding='unicode_escape', sep=';', index_col=0, decimal=',')

df = df[STATION_COLUMNS]
df = df[df['Water Type'].isin(WATER_TYPES)]

df.dropna(inplace=True)

stations = df

stations.head()
```

```python
DISSOLVED_GAS_COLUMNS = ['GEMS.Station.Number', 'Depth', 'Value', 'Unit', 'Datetime']
PARAMETER_CODES = ['O2-Dis', 'O2-Dis-Sat']

df = pd.read_csv('data/GFQA_v2/Dissolved_Gas.csv', sep=';',
                 dtype={
                     'Depth': float,
                     'Analysis.Method.Code': str,
                     'Value.Flags': str,
                     'Value': float,
                 })

df['Datetime'] = pd.to_datetime(
    df['Sample.Date'] + ' ' + df['Sample.Time'], format='%Y-%m-%d %H:%M')

df = df[df['Parameter.Code'].isin(PARAMETER_CODES)]

df = df[DISSOLVED_GAS_COLUMNS]

o2_dis = df.pivot_table(index=['GEMS.Station.Number', 'Depth',
                        'Datetime'], columns='Unit', values='Value').reset_index()

o2_dis = o2_dis.merge(stations, left_on='GEMS.Station.Number',
             right_index=True).reset_index(drop=True)

o2_dis.to_csv('data/o2_dis.csv', index=False)
o2_dis.head()
```
